## Making a Neutral Account

```python
stripe_public = os.getenv('stripe_public')
stripe_private = os.getenv('stripe_private')
stripe.api_key=stripe_private
```

```python
stripe.Account.create(
    type='custom',
    country='US',
    email='andrew.stiner@yahoo.com',
    capabilities={
        "card_payments":{"requested":True},
        "transfers":{"requested":True},
    },
    business_type='non_profit'
)
```

## Creating People Associated With This Account

```python
name_list = ['Andrew', 'Chris', 'Alan', 'Gabe']

for name in name_list:
    stripe.Account.create_person(
        "acct_1HILPbIW2XJXscel",
        first_name=name,
    )
```

```python
#Listing the people created in the cell above to get their account and person ID
stripe.Account.list_persons(
    "acct_1HILPbIW2XJXscel",
    limit=4,
)
```

## Making Us All Customers So That We Can Pay and Receive From The Neutral Account

```python
stripe.Customer.create(
    description="This test customer account will be for Andrew",
    name="Andrew"
)
stripe.Customer.create(
    description="This test customer account will be for Chris",
    name="Chris"
)
stripe.Customer.create(
    description="This test customer account will be for Alan",
    name="Alan"
)
stripe.Customer.create(
    description="This test customer account will be for Gabe",
    name="Gabe"
)
```

```python
stripe.Customer.list(limit=4)
```

## Modifying the People Created Eariler To Be Attributed Legal Authorization And Responsibility Over The Account

```python
person_id = ["person_Hs5cGZATysVXkf", "person_Hs5ch8SyyDRPmz", "person_Hs5cVvFIZNL6U0", "person_Hs5cGWOuQRRjtV"]

for person in person_id:
    stripe.Account.modify_person(
        "acct_1HILPbIW2XJXscel",
        person,
        relationship={"executive":True}
    )
```

## Modifying Account To Show The Change Made Above

```python
stripe.Account.modify(
    "acct_1HILPbIW2XJXscel",
    company={
        "executives_provided":True,
    }
)
```

## Creating Sources For Everyone
This sets up our payment methods

```python
email_list = ['Andrew@test.com', 'Chris@test.com', 'Alan@test.com', 'Gabe@test.com']
for email in email_list:
    source = stripe.Source.create(
        type='ach_credit_transfer',
        currency='usd',
        owner={
            'email':email
        }
    )
    print(source)
```
